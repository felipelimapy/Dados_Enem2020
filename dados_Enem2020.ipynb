```python
import pandas as pd
```

```python
df= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/MICRODADOS_ENEM_2020.csv", encoding='ISO-8859-1', sep=";")
```

```python
dados_selecionados=df[['TP_COR_RACA','TP_SEXO','TP_ESCOLA', 'NU_NOTA_MT', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]

#selecionando apenas colunas que quero trabalhar
```

```python
dados_selecionados['NOTAS_TOTAIS']= 0
#criação de uma nova coluna pra armazenarmos a nota total de cada candidato
```

```python
dados_selecionados['NOTAS_TOTAIS'] = dados_selecionados.iloc[:, 3:].sum(axis=1)/5
```

```python
dados_selecionados
```

```python
dados_selecionados['TP_ESCOLA']=dados_selecionados['TP_ESCOLA'].astype(object)
dados_selecionados['TP_COR_RACA']=dados_selecionados['TP_COR_RACA'].astype(object)
#tranformando números em textos
```

```python
dados_selecionados['TP_ESCOLA']=dados_selecionados['TP_ESCOLA'].replace([1,2,3,4],['nao respondeu', 'publica', 'privada', 'exterior'])
dados_selecionados['TP_COR_RACA']=dados_selecionados['TP_COR_RACA'].replace([0,1,2,3,4,5], ['nao declarou', 'branca', 'preta', 'parda', 'amarela', 'indigena'])
#atribuindo textos nos números
```

```python
dados_selecionados=dados_selecionados.dropna()
#eliminando linhas sem valores
```

```python
dados_selecionados
```

1-Qual a escola com a maior média de notas?

```python
ordenados=dados_selecionados.groupby(by="TP_ESCOLA").mean().round(2)

ordenados.sort_values("NOTAS_TOTAIS",ascending=False)
#ranking das escolas com maiores notas
```

2-Qual o aluno com a maior média de notas e o valor dessa média?

```python
dados_selecionados.iloc[dados_selecionados['NOTAS_TOTAIS'].argmax()]
```
